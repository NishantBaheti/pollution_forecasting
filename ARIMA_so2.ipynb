```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
%matplotlib inline
```

```python
df = pd.read_excel("Delhi.xlsx")
```

```python
for i in range(319):
    df['Date'][i]=df['Date'][i][0:10]
```

```python
df["Date"] = pd.to_datetime(df["Date"],format="%d-%m-%Y")
```

```python
df['Date']
```

```python
df_so2 = pd.DataFrame(data =df , columns=['Date','SO2'])
```

```python
df_so2
```

```python
plt.plot(df_so2['Date'],df_so2['SO2'])
```

```python
indexeddataset = df_so2.set_index(['Date'])
```

```python
#detecting rolling statistics
rollmean = indexeddataset.rolling(window =30).mean()
rollstd = indexeddataset.rolling(window =30).std()

print(rollmean,rollstd)
```

```python
#plotting rolling statistics
orig = plt.plot(indexeddataset ,color ='blue', label ='original')
mean = plt.plot(rollmean, color='red', label='Rolling mean')
std = plt.plot(rollstd, color='black', label='Rolling Standard')
plt.legend(loc='best')
plt.title("rolling mean and standard")
plt.show(block=False)
```

```python
from statsmodels.tsa.stattools import adfuller

print("Results of dickey-fuller-test")
dftest = adfuller(indexeddataset['SO2'],autolag='AIC')

dfoutput = pd.Series(dftest[0:4], index=['test statistics','p-value','#lagsused','number of obeservations used'])
for key,value in dftest[4].items():
    dfoutput['critical value (%s)'%key] = value
    
print(dfoutput)
```

```python
#results of dickey fuller test shows that data is not stationary
# p-value should be around 0.5
# test statistics < critical value
```

```python
indexeddataset_logscale = np.log(indexeddataset)
plt.plot(indexeddataset_logscale)
```

```python
movingaverage = indexeddataset_logscale.rolling(window=30).mean()
movingstd = indexeddataset_logscale.rolling(window=30).std()
plt.plot(indexeddataset_logscale)
plt.plot(movingaverage,color='red')
```

```python
datasetlogscaleminusmovingaverage = indexeddataset_logscale - movingaverage
datasetlogscaleminusmovingaverage.head(30)

datasetlogscaleminusmovingaverage.dropna(inplace=True)
datasetlogscaleminusmovingaverage.head(30)
```

```python
from statsmodels.tsa.stattools import adfuller
def test_stationary(timeseries):
    #determine rolling stats
    movingaverage = timeseries.rolling(window=30).mean()
    movingstd = timeseries.rolling(window=30).std()
    
    #plotting rolling statistics
    orig = plt.plot(timeseries ,color ='blue', label ='original')
    mean = plt.plot(movingaverage, color='red', label='Rolling mean')
    std = plt.plot(movingstd, color='black', label='Rolling Standard')
    plt.legend(loc='best')
    plt.title("rolling mean and standard")
    plt.show(block=False)
    
    #perform dickey fuller test
    print("Results of dickey-fuller-test")
    dftest = adfuller(timeseries['SO2'],autolag='AIC')

    dfoutput = pd.Series(dftest[0:4], index=['test statistics','p-value','#lagsused','number of obeservations used'])
    for key,value in dftest[4].items():
        dfoutput['critical value (%s)'%key] = value
    
    print(dfoutput)
    
```

```python
test_stationary(datasetlogscaleminusmovingaverage)
```

```python
exponentialDecayweightedaverage = indexeddataset_logscale.ewm(halflife=12, min_periods=0,adjust=True).mean()
plt.plot(indexeddataset_logscale)
plt.plot(exponentialDecayweightedaverage,color='Red')
```

```python
datasetlogscaleminusmovingexponantialdecayaverage = indexeddataset_logscale - exponentialDecayweightedaverage                
test_stationary(datasetlogscaleminusmovingexponantialdecayaverage)
```

```python
datasetlogdiffshifitng = indexeddataset_logscale - indexeddataset_logscale.shift()             
plt.plot(datasetlogdiffshifitng)
```

```python
datasetlogdiffshifitng.dropna(inplace=True)
test_stationary(datasetlogdiffshifitng)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(indexeddataset_logscale)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(indexeddataset_logscale,label ='original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='trend')
plt.legend(loc = 'best')
plt.subplot(413)
plt.plot(seasonal,label="seasonality")
plt.legend(loc ='best')
plt.subplot(414)
plt.plot(residual,label='residual')
plt.legend(loc ='best')
plt.tight_layout()

```

```python
decomposedlogdata = residual
decomposedlogdata.dropna(inplace= True)
test_stationary(decomposedlogdata)
```

```python
from statsmodels.tsa.stattools import acf,pacf

lag_acf = acf(datasetlogdiffshifitng, nlags=20)
lag_pacf = pacf(datasetlogdiffshifitng, nlags=20 ,method='ols')




#plot acf
plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(datasetlogdiffshifitng)),linestyle='--',color='gray')  
plt.axhline(y=1.96/np.sqrt(len(datasetlogdiffshifitng)),linestyle='--',color='gray')
plt.title('Autocorrelation function')


#plot pacf
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(datasetlogdiffshifitng)),linestyle='--',color='gray')  
plt.axhline(y=1.96/np.sqrt(len(datasetlogdiffshifitng)),linestyle='--',color='gray')
plt.title('partial Autocorrelation function')
plt.tight_layout()
```

```python
from statsmodels.tsa.arima_model import ARIMA

#AR model 
model = ARIMA(indexeddataset_logscale,order=(1,1,1))
results_ar = model.fit(disp=-1)
plt.plot(datasetlogdiffshifitng)
plt.plot(results_ar.fittedvalues,color='red')
plt.title("RSS: %.4f"% sum((results_ar.fittedvalues-datasetlogdiffshifitng["SO2"])**2))
print("plotting AR model")
```

```python
from statsmodels.tsa.arima_model import ARIMA
from math import sqrt

#AR model 
model = ARIMA(indexeddataset_logscale,order=(1,1,1))
results_ar = model.fit(disp=-1)
plt.plot(datasetlogdiffshifitng)
plt.plot(results_ar.fittedvalues,color='red')
plt.title("RSS: %.4f"% sum((results_ar.fittedvalues-datasetlogdiffshifitng["SO2"])**2))
print("plotting AR model")
```

```python
predictions_arima_diff = pd.Series(results_ar.fittedvalues, copy=True)
print(predictions_arima_diff.head())
```

```python
predictions_arima_diff_cumsum = predictions_arima_diff.cumsum()
print(predictions_arima_diff_cumsum.head())
```

```python
predictions_arima_log = pd.Series(indexeddataset_logscale["SO2"].ix[0],indexeddataset_logscale.index)
predictions_arima_log = predictions_arima_log.add(predictions_arima_diff_cumsum,fill_value=0)
predictions_arima_log.head()
```

```python
predictions_arima = np.exp(predictions_arima_log)
plt.plot(indexeddataset)
plt.plot(predictions_arima,color="red")
```
